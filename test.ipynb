```python
from typing import List
import re
import ast
import json
class FunctionCall:
    def __init__(self, name: str, arguments: str):
        self.name = name
        self.arguments = arguments

def parse_function_calls(calls: List[str]) -> List[FunctionCall]:
    results = []
    for call in calls:
        # 匹配函数名和参数字符串
        match = re.match(r"(\w+(?:-\w+)*)\((.*)\)", call)
        if not match:
            continue
        func_name, args_str = match.groups()
        # 把类似 key='value' 的参数转化为 dict
        args = {}
        for arg in re.findall(r"(\w+)=['\"]([^'\"]+)['\"]", args_str):
            k, v = arg
            args[k] = v
        results.append(FunctionCall(name=func_name, arguments=json.dumps(args)))
    return results
```
